```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python

df = sns.load_dataset('Downloads/diamonds')
df
```

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

```python
sns.set({'figure.figsize':(13,13)})
palette = "ch:r=-.2,d=.3_r"
#plt.xlim(0,5)
sns.scatterplot(x='carat', y='price', hue='clarity', size='depth', palette=palette, data=df)

plt.title("Scatterplot of Diamond Carat vs. Price with Cut as Hue")
plt.xlabel("Carat")
plt.ylabel("Price")
plt.legend(title='Cut')
plt.show()
```

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price serta markers =  ['p', 'o', 's', 'D', 'X'] sehingga dihasilkan output sesuai expected.

```python
df = df.sample(n=300, random_state=123)

#code here
    
sns.pairplot(df, hue='cut', vars=["carat", "depth", "table", "price"], diag_kind='kde', markers=["p", "o", "s", "D", "X"], corner=True);
```
