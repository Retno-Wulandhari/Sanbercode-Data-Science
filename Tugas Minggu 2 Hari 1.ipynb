<h2>Soal 3: DataFrame vs Series</h2>

- Apa perbedaan Series dengan list dan dictionary
- Apa perbedaan dataframe dan series
- Buatlah suatu data series

Jawab: 
- Series memiliki persamaan dengan list yaitu dapat dilakukan metode slicing, yaitu mengakses/indexing pada object series tersebut. Juga series memiliki persamaan dengan dictionary yaitu index yang dimiliki series berguna seperti keys pada dictionary. Sesuai perbedaan data dari list dan dictionary, series dengan list data memiliki urutan dan bisa dirubah sedangkan series dengan dictionary koleksi pasangan key yang digunakan untuk mengakses data.
  
- Perbedaan dari Series dan DataFrame ini terletak pada strukturnya dan juga attribute dan method-method yang mereka miliki, untuk perbedaan strukturnya adalah sebagai berikut :

Series adalah objek 1 dimensi
DataFrame adalah suatu objek 2 dimensi, mirip seperti tabel

```python
# code here
x = pd.Series([80,70,84], index=['ana', 'ali', 'cici'])
print(x)
```

<h2>Soal 5: Deskriptif Statistik</h2>

Jelaskan apa itu deskriptis statistik!


Statistik deskriptif mencakup statistik yang merangkum tendensi sentral, penyebaran, dan bentuk distribusi kumpulan data.

Statistik deskriptif merupakan proses analisis statistik yang fokus kepada manejemen, penyajian, dan klasifikasi data. Dengan proses ini, data yang disajikan akan menjadi lebih menarik lebih mudah dipahami, dan mampu memberikan makna lebih bagi pengguna data.

Statistik deskriptif memberikan deskripsi data secara ringkas dari bagaimana bentuk distribusi datanya, sampai ukuran penyebaran dari data tersebut

<h2>Soal 6: Practice Deskriptif Statistik</h2>

- Gunakan suatu metode dari pandas untuk mengetahui deskriptif statistik dari suatu data
- Berapa nilai rata-rata dari column price
- Berapa nilai standar deviasi dari column depth
- Berapa nilai maximum dari column carat

```python
# code here
df = pd.read_csv('diamonds.csv')
df.describe()
```

```python
rerata = sum(df['price']) / len(df['price'])
print(rerata)

print(np.std(df['depth'], ddof=1))

print(max(df['carat']))
```

Jawab:
- Metode yang di gunakan adalah describe() hasil tampilan seperti diatas
- Nilai rata-rata dari column price  :  3932.799721913237
- Nilai standar deviasi dari column depth : 1.432621318833661
- Nilai maximum dari column carat : 5.01
