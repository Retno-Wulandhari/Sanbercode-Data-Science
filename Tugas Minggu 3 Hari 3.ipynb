<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
#code here
#fig = plt.figure()
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(20,5))
ax[0].plot(x, tan, color='blue', label='tanX')
ax[0].axis("off")
ax[1].plot(x, sin, color='red', label='sinX')
ax[1].axis("off")
ax[2].plot(x, cos, color='green', label='cosX')
ax[2].axis("off")

x1, y1 = ax[0].get_legend_handles_labels()
x2, y2 = ax[1].get_legend_handles_labels()
x3, y3 = ax[2].get_legend_handles_labels()

fig.legend(x1 + x2 + x3, y1 + y2 + y3, loc='upper center', fontsize='small')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here
fig, ax = plt.subplots(figsize=(15,12))

ax.plot(x, y)
ax.axis("off")

ins1 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3]) #kanan bawah
ins1.plot(x, y2, color='red')
ins1.plot(x, y, color='blue')

ins2 = ax.inset_axes([0.9, -0.1, 0.3, 0.3]) #kiri bawah
ins2.plot(x, y3, color='green')
ins2.plot(x, y, color='blue')

ins3 = ax.inset_axes([0.9, 0.9, 0.3, 0.3]) #kanan atas
ins3.plot(x, y4, color='black')
ins3.plot(x, y, color='blue')

ins4 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3]) #kiri atas
ins4.plot(x, y5, color='yellow')
ins4.plot(x, y, color='blue')

ins1.set_title('Grafik CosX^2')
ins2.set_title('Grafik CosX^3')
ins3.set_title('Grafik CosX^4')
ins4.set_title('Grafik CosX^5')

plt.show()
```
